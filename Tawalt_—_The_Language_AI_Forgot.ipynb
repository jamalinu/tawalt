```python
from rapidfuzz import process

diccionario_palabras = list(respuestas_tarifit.keys())

def buscar_palabra(mensaje):
    palabra = mensaje.lower().strip()

    # Búsqueda exacta primero
    if palabra in respuestas_tarifit:
        return respuestas_tarifit[palabra]

    # Búsqueda aproximada
    resultado = process.extractOne(palabra, diccionario_palabras, score_cutoff=80)
    if resultado:
        palabra_encontrada = resultado[0]
        return f"(¿Quisiste decir '{palabra_encontrada}'?) {respuestas_tarifit[palabra_encontrada]}"

    return None

# Probamos
print(buscar_palabra("tanemirt"))   # error tipográfico
print(buscar_palabra("azul"))       # correcto
print(buscar_palabra("hello"))      # desconocido
```
